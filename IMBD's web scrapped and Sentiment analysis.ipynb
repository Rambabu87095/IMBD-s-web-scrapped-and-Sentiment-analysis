```python
from bs4 import BeautifulSoup
import requests
import pandas as pd 
```

```python
url='https://www.imdb.com/list/ls095374765/?sort=list_order,asc&st_dt=&mode=detail&page=1'
```

```python
res=requests.get(url)
content=BeautifulSoup(res.content,'html.parser')
```

```python
movie=content.find_all('div',{'class':'lister-item-content'})
movie_name=content.find_all('h3',{'class':'lister-item-header'})
```

```python
Movie_name=[]
for i in movie_name:
    Movie_name.append(i.a.text)
Movie_name
```

```python
yr=content.find_all('span',{'class':'lister-item-year text-muted unbold'})
```

```python
yea=[]
a=''
for i in yr:
    yea.append(i.text)
for i in yea:
    for j in i:
        for k in j:
            if k.isnumeric():
                a=a+k
#a
```

```python
b=''
for j,i in enumerate(a):
    b=b+i
    if j==0:
        continue
    elif j==1:
        continue
    elif j==2:
        continue
    elif j==3:
        b=b+','
    elif j%4==1:
        continue
    elif j%2==1:
        b=b+','
Year=b.split(',')
Year=Year[:-1]
Year.insert(40,'na')
Year
```

```python
import pandas as pd
```

```python
review=content.find_all('p',{'class':''})
view=[]
Review=[]
for i in review:
    view.append(i.text)
for i in view:
    i=i[1:]
    Review.append(i)
Review
```

```python
genre=content.find_all('span',{'class':'genre'})
Gen=[]
Genre=[]
for i in genre:
    Gen.append(i.text)
for i in Gen:
    Genre.append(i[1:])
Genre
```

```python
rat=content.find_all('span',{'class':'ipl-rating-star__rating'})
rate=[]
Rating=[]
for i in rat:
    rate.append(i.text)
for i,j in enumerate(rate):
    if i%23==0:
        Rating.append(j)
Rating.insert(40,'na')
Rating
```

```python
lent=content.find_all('span',{'class':'runtime'})
Movie_time=[]
for i in lent:
    Movie_time.append(i.text)
Movie_time.insert(40,'na')
Movie_time
```

```python
di=content.find_all('p',class_='text-muted text-small')   
```

```python
c=[]
for i in di:
    c.append(i.find_all('a'))
#c
dire=[]
for i in c:
    if i==[]:
        c.remove(i)
    else:
        dire.append(i)
#dire    
dd=[]
for i in c:
    for j in i:
        dd.append(j.text)
#dd
```

```python
Director=[]
actor=[]
for i,j in enumerate(dd):
    if i%5==0:
        Director.append(j)
    else:
        actor.append(j)
Director.insert(40,'na')
Director
```

```python
u=[]
l=''
for i,j in enumerate(actor):
    if i%4==0:
        l='@'+j
    else:
        l=l+','+j
    u.append(l)
#u  
```

```python
act=[]
for i,j in enumerate(u):
    if i%4==3:
        act.append(j)
        
acto=[]
for i in act:
    acto.append(i.split('@'))
    
Actor=[]
for i in acto:
    for j in i:
        if j!='':
            Actor.append(j)
            
Actor.insert(40,'na')
Actor    
```

```python
data=pd.DataFrame({'Movie_name':Movie_name,'Year':Year,'Review':Review,'Genre':Genre,'Rating':Rating,'Movie_time':Movie_time,'Director':Director,'Actor':Actor})
data
```

### stop words

```python
#stop words
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
#stop_words
data['Review']
data['remove stop-words'] = data['Review'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
data['remove stop-words']
```

### sentiment analysis

```python
from nltk.sentiment import SentimentIntensityAnalyzer
sia=SentimentIntensityAnalyzer()
```

```python
r1=[]
for i in range(len(data['remove stop-words'])):
    r1.append(sia.polarity_scores(data['remove stop-words'][i]))
negative=[]
for i in r1:
    negative.append(i['neg'])
negative
```

```python
r1=[]
for i in range(len(data['remove stop-words'])):
    r1.append(sia.polarity_scores(data['remove stop-words'][i]))
neutral=[]
for i in r1:
    neutral.append(i['neu'])
neutral
```

```python
r1=[]
for i in range(len(data['remove stop-words'])):
    r1.append(sia.polarity_scores(data['remove stop-words'][i]))
positive=[]
for i in r1:
    positive.append(i['pos'])
positive
```

```python
r1=[]
for i in range(len(data['remove stop-words'])):
    r1.append(sia.polarity_scores(data['remove stop-words'][i]))
compound=[]
for i in r1:
    compound.append(i['compound'])
compound
```

```python
data['Negative-remove stop-words']=negative
data['Neutral-remove stop-words']=neutral
data['Positive-remove stop-words']=positive
data['Compound-remove stop-words']=compound
```
